# Scatter

## 1. Import matplotlib and numpy

```python
import matplotlib.pyplot as plt 
import numpy as np
from matplotlib import font_manager


plt.rcParams['mathtext.fontset'] = 'stix' # 'stix' | 'cm'
plt.rcParams['font.size'] = 12

plt.rcParams['xtick.direction'] = 'in'
plt.rcParams['ytick.direction'] = 'in'
```

## 2. Prepare Data

```python
np.random.seed(341126)

x = 1.5*np.random.random(100) + 0.2
y1 = x**2 + 0.5*np.random.random(100)
y2 = x**(0.5) + 0.25*np.random.random(100)
```

## 3. Call Plot API

```python
fig, ax = plt.subplots(figsize=(3.5, 3))

_ = ax.scatter(x, y1, color='r', marker='s', s=24, label='data: y1')
_ = ax.scatter(x, y2, color='blue', marker='o', s=32, alpha=0.5, label='data: y2')
_ = ax.legend(frameon=False, loc=0, title='')


ax.set_xlabel("x/unit")
ax.set_ylabel("y/unit")

# plt.savefig('demo.png', dpi=300, bbox_inches='tight')

plt.show()
```
