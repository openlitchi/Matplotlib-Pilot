# Histogram statistical

## 1. Import matplotlib and numpy

```python
import matplotlib.pyplot as plt 
import numpy as np
from matplotlib import font_manager


plt.rcParams['font.size'] = 12

plt.rcParams['xtick.direction'] = 'in'
plt.rcParams['ytick.direction'] = 'in'
```

## 2. Prepare Data

```python
np.random.seed(341126)
N = 1<<14

# 生成两个标准正态分布的随机数
data1 = np.random.randn(N) - 3  # 第一个峰
data2 = 1.5*np.random.randn(N) + 2  # 第二个峰
```

## 3. Call Plot API

```python
fig, ax = plt.subplots(figsize=(3.5, 3))

_ = ax.hist(data1, bins=32, density=True, alpha=0.75, label="data 1")
_ = ax.hist(data2, bins=32, density=True, alpha=0.75, label="data 2")

ax.set_xlabel("x/unit")
ax.set_ylabel("number")

_ = ax.legend(frameon=False, loc=0, title='')

# plt.savefig('demo.png', dpi=300, bbox_inches='tight')

plt.show()
```
