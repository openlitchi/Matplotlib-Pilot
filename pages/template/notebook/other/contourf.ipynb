# Contour Plot

## 1. Import matplotlib and numpy

```python
import matplotlib.pyplot as plt 
import numpy as np
from matplotlib import font_manager


plt.rcParams['font.size'] = 10

plt.rcParams['xtick.direction'] = 'in'
plt.rcParams['ytick.direction'] = 'in'
```

## 2. Prepare Data

```python
x,y = np.meshgrid(np.arange(-32, 32), np.arange(-32, 32))

r = np.sqrt(x**2 + y**2)
array2d = np.sin(0.06*np.pi*r)
```

## 3. Call Plot API

```python
fig, ax = plt.subplots(figsize=(5,4))

levels = [-1,-0.5,0,0.5,1]

CS1 = ax.contourf(array2d, 
                 levels=levels, 
                 cmap='coolwarm')

CS2 = ax.contour(array2d, 
            levels=levels,
            colors='k',
            linewidths=1.5,
            linestyles='-')

ax.clabel(CS2, fmt='%2.1f', colors='k', fontsize=10)

fig.colorbar(CS1, 
             ax=ax, 
             label='value / unit', 
             pad=0.05,
             shrink=1.00)

# aspect = dy/dx
ax.set_aspect(aspect = 1.0)

# plt.savefig('demo.png', dpi=300, bbox_inches='tight')

plt.show()
```
