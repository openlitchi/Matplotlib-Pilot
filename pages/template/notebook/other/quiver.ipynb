# Vector arrow diagram

## 1. Import matplotlib and numpy

```python
import matplotlib.pyplot as plt 
import numpy as np
from matplotlib import font_manager


plt.rcParams['font.size'] = 10

plt.rcParams['xtick.direction'] = 'in'
plt.rcParams['ytick.direction'] = 'in'
```

## 2. Prepare Data

```python
X, Y = np.meshgrid(np.arange(0, 2 * np.pi, .2), np.arange(0, 2 * np.pi, .2))
U = np.cos(X)
V = np.sin(Y)
```

## 3. Call Plot API

```python
fig, ax = plt.subplots(figsize=(3.5, 3))
ax.set_title("quiver")


M = np.hypot(U, V)
quiver = ax.quiver(X[::2, ::2], Y[::2, ::2], U[::2, ::2], V[::2, ::2], M[::2, ::2],
                    cmap='coolwarm')
# _ = ax.scatter(X[::2, ::2], Y[::2, ::2], color='r', s=5)

# 隐藏刻度线
ax.tick_params(axis='both', which='both', length=0)

# 隐藏刻度标签
ax.set_xticklabels([])
ax.set_yticklabels([])

# 添加colorbar
cbar = fig.colorbar(quiver, ax=ax, boundaries=np.linspace(0, 1.5, 151))

cbar.set_label("Velocity (m/s)")

# 设置colorbar的刻度
cbar.set_ticks([0.0, 0.3, 0.6, 0.9, 1.2, 1.5])

# plt.savefig('demo.png', dpi=300, bbox_inches='tight')

plt.show()
```
