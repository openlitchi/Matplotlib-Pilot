# Bar with nummber labels

## 1. Import matplotlib and numpy

```python
import matplotlib.pyplot as plt 
import numpy as np
from matplotlib import font_manager


plt.rcParams['font.size'] = 12

plt.rcParams['xtick.direction'] = 'in'
plt.rcParams['ytick.direction'] = 'in'
```

## 2. Prepare Data

```python
labels = ['litchi', 'banana', 'grape', 'chestnut', 'kiwifruit', 'citrus', 'peach']
mass = [20, 29, 30, 35, 27, 22, 34]
```

## 3. Call Plot API

```python
fig, ax = plt.subplots(figsize=(4,4))
rects = ax.bar(labels, mass, width=0.9, color = 'tomato', alpha = 0.85, label='Fruit')

# ylabel
ax.set_ylabel('Scores')

# 获取当前的x轴刻度标签
xtick_labels = ax.get_xticklabels()
# 设置x轴刻度标签旋转45度
ax.set_xticklabels(xtick_labels, rotation=45)

# title
ax.set_title('Scores by group and gender', y=-0.36)
# ax.legend(loc=2)

ax.bar_label(rects, padding=3)
ax.set_ylim([0, 40])

fig.tight_layout()

# plt.savefig('demo.png', dpi=300, bbox_inches='tight')
plt.savefig('./cover/bar_num.png', dpi=100, bbox_inches='tight')

plt.show()
```
