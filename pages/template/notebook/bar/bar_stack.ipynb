# Stack bar

## 1. Import matplotlib and numpy

```python
import matplotlib.pyplot as plt 
import numpy as np
from matplotlib import font_manager


plt.rcParams['font.size'] = 12

plt.rcParams['xtick.direction'] = 'in'
plt.rcParams['ytick.direction'] = 'in'
```

## 2. Prepare Data

```python
item = ('U', 'V', 'W', 'X')
data = {
    'A': np.array([73, 34, 61, 45]),
    'B': np.array([47, 44, 58, 37]),
    'C': np.array([53, 37, 57, 39])
}
```

## 3. Call Plot API

```python
fig, ax = plt.subplots(figsize=(3.5, 3))

width = 0.6  # the width of the bars: can also be len(x) sequence

bottom = np.zeros(4)

colors = ['lightsalmon','lightgreen','dodgerblue']
i = 0
for sex, sex_count in data.items():
    p = ax.bar(item, sex_count, width, label=sex, bottom=bottom, color=colors[i])
    bottom += sex_count

    ax.bar_label(p, label_type='center')
    i += 1

ax.set_title('Title Here')
ax.legend(frameon=False, ncols=3)
ax.set_ylim(0,220)

# ax.set_xlabel("x label")
# ax.set_ylabel("y / unit")

# plt.savefig('demo.png', dpi=300, bbox_inches='tight')

plt.show()
```
