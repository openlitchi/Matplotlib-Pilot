# No marker line

## 1. Import matplotlib and numpy

```python
import matplotlib.pyplot as plt 
import numpy as np
from matplotlib import font_manager


plt.rcParams['font.size'] = 10

plt.rcParams['xtick.direction'] = 'in'
plt.rcParams['ytick.direction'] = 'in'
```

## 2. Prepare Data

```python
x = [1, 2, 3, 4, 5, 6, 7]
y1 = [0.31, 0.37, 0.42, 0.41, 0.35, 0.32, 0.30]
y2 = [0.27, 0.31, 0.38, 0.44, 0.45, 0.39, 0.36]
```

## 3. Call Plot API

```python
fig, ax = plt.subplots(figsize=(3.5, 3))

_ = ax.plot(x, y1, color='r', ls='-', label='data: y1')
_ = ax.plot(x, y2, color='b', ls='-', label='data y2')
_ = ax.legend(frameon=False, loc=0, title='')

ax.set_xlabel("x/unit")
ax.set_ylabel("y/unit")

ax.set_xticks(x)
ax.set_yticks([0.3, 0.35, 0.40])

# plt.savefig('demo.png', dpi=300, bbox_inches='tight')

plt.show()
```
