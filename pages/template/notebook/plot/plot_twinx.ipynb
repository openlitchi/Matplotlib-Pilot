# Two Y Axis

## 1. Import matplotlib and numpy

```python
import matplotlib.pyplot as plt 
import numpy as np
from matplotlib import font_manager


plt.rcParams['font.size'] = 10

plt.rcParams['xtick.direction'] = 'in'
plt.rcParams['ytick.direction'] = 'in'
```

## 2. Prepare Data

```python
x = [1, 2, 3, 4, 5, 6, 7]
y1 = [0.31, 0.37, 0.42, 0.41, 0.35, 0.32, 0.30]
y2 = [2.7, 3.1, 3.88, 4.4, 4.5, 3.9, 3.6]
```

## 3. Call Plot API

```python
fig, ax = plt.subplots(figsize=(3.5, 3))

color_left = 'r'
color_right = 'b'

_ = ax.plot(x, y1, color=color_left, marker='s', ls='-', label='data: y1')
ax.set_xticks(x)
ax.set_xlabel("x/unit")
ax.set_ylim(0.29, 0.45)
ax.set_yticks([0.3, 0.35, 0.40])
ax.set_ylabel("y1/unit", color=color_left)

ax_right = ax.twinx()
_ = ax_right.plot(x, y2, color=color_right, marker='o', ls='-', label='data y2')
ax_right.set_ylim(2.4, 5.0)
ax_right.set_yticks([2.5, 3.5, 4.5])
ax_right.set_ylabel("y2/unit", color=color_right)

# 设置刻度颜色
ax.tick_params(axis='y', colors=color_left) 
ax_right.tick_params(axis='y', colors=color_right) 

# 设置边框颜色
ax_right.spines['left'].set_color(color_left)
ax_right.spines['right'].set_color(color_right)

# 设置legend
_ = ax.legend(frameon=False, loc="upper left", title='')
_ = ax_right.legend(frameon=False, loc="upper right", title='')

# 保存图片
# plt.savefig('demo.png', dpi=300, bbox_inches='tight')

plt.show()
```
