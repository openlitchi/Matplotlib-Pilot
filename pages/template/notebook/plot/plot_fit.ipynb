# Fitted Line

## 1. Import matplotlib and numpy

```python
import matplotlib.pyplot as plt 
import numpy as np
from matplotlib import font_manager


plt.rcParams['font.size'] = 12

plt.rcParams['xtick.direction'] = 'in'
plt.rcParams['ytick.direction'] = 'in'
```

## 2. Prepare Data

```python
x = np.linspace(0, 10, 11)
y = [3.9, 4.4, 8.8, 10.3, 11.2, 13.1, 14.1,  13.9, 13.9, 14.1, 12.5]

```

**Fit Data**

```python
a, b, c = np.polyfit(x, y, deg=2)

N = 51

x_fit = np.linspace(0, 10, N)
y_fit = a * x_fit**2 + b * x_fit +c
```

## 3. Call Plot API

```python
fig, ax = plt.subplots(figsize=(3.5, 3))

ax.plot(x_fit, y_fit, c='tomato', lw=2, label='fit data')
ax.scatter(x, y, marker='o', color='coral', label='origin data')
_ = ax.legend(frameon=False, loc=0, title='legend')

ax.set_xlabel("x/unit")
ax.set_ylabel("y/unit")

# plt.savefig('demo.png', dpi=300, bbox_inches='tight')

plt.show()
```
