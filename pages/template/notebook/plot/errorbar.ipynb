# Line with errorbar

## 1. Import matplotlib and numpy

```python
import matplotlib.pyplot as plt 
import numpy as np
from matplotlib import font_manager


plt.rcParams['font.size'] = 10

plt.rcParams['xtick.direction'] = 'in'
plt.rcParams['ytick.direction'] = 'in'
```

## 2. Prepare Data

```python
x = [1, 2, 3, 4, 5, 6, 7]
y1 = [0.31, 0.37, 0.42, 0.41, 0.35, 0.32, 0.30]
y2 = [0.27, 0.31, 0.38, 0.44, 0.45, 0.39, 0.36]

np.random.seed(341126)
err1 = 0.02*np.random.random(len(y1))
err2 = 0.02*np.random.random(len(y1))
```

## 3. Call Plot API

```python
fig, ax = plt.subplots(figsize=(3.5,3))  

ax.errorbar(x, y1, err1, 
    ls='-',  c='orangered', marker='s', 
    capsize=4, alpha=0.75, label='data: y1')

ax.errorbar(x, y2, err2, 
    ls='-',  c='red', marker='P', 
    capsize=4, alpha=0.75, label='data: y2')

ax.legend(frameon=False, loc=0)
ax.set_xlabel('x / m')
ax.set_ylabel(r'$y_1$ and $y_2$', labelpad=2, y=0.5)

# plt.savefig('demo.png', dpi=300, bbox_inches='tight')

plt.show()
```
