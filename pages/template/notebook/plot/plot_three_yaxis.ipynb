# Three Y Axis

## 1. Import matplotlib and numpy

```python
import matplotlib.pyplot as plt 
import numpy as np
from matplotlib import font_manager


plt.rcParams['font.size'] = 10

plt.rcParams['xtick.direction'] = 'in'
plt.rcParams['ytick.direction'] = 'in'
```

## 2. Prepare Data

```python
x = [1, 2, 3, 4, 5, 6, 7]
y1 = [0.31, 0.37, 0.42, 0.41, 0.35, 0.32, 0.30]
y2 = [2.7, 3.1, 3.88, 4.4, 4.5, 3.9, 3.6]
y3 = [2.7, 31, 388, 1.4e3, 2.5e4, 2.9e5, 1.1e6]
```

## 3. Call Plot API

```python
fig, ax1 = plt.subplots(figsize=(3.5, 3))

color1 = 'r'
color2 = 'b'
color3 = 'g'

_ = ax1.plot(x, y1, color=color1, marker='s', ls='-', label='data: y1')
ax1.set_xticks(x)
ax1.set_xlabel("x/unit")
ax1.set_ylim(0.29, 0.45)
ax1.set_yticks([0.3, 0.35, 0.40])
ax1.set_ylabel("y1/unit", color=color1)

ax2 = ax1.twinx()
_ = ax2.plot(x, y2, color=color2, marker='o', ls='-', label='data: y2')
ax2.set_ylim(2.4, 5.0)
ax2.set_yticks([2.5, 3.5, 4.5])
ax2.set_ylabel("y2/unit", color=color2)

ax3 = ax1.twinx()
_ = ax3.plot(x, y3, color=color3, marker='o', ls='-', label='data: y3')
ax3.set_yscale('log')
ax3.set_ylim(1.0, 6.0e6)
# ax3.set_yticks([2.5, 3.5, 4.5])
ax3.set_ylabel("y3/unit", color=color3)
# 移动第三个Y轴到右边，以避免覆盖第二个Y轴
ax3.spines['right'].set_position(('outward', 40))

# 设置刻度颜色
ax1.tick_params(axis='y', colors=color1) 
ax2.tick_params(axis='y', colors=color2) 
ax3.tick_params(axis='y', colors=color3) 

# 设置边框颜色
ax1.spines['left'].set_color(color1)
ax2.spines['right'].set_color(color2)
ax3.spines['right'].set_color(color3)

# 设置legend
_ = ax1.legend(frameon=False, loc="upper left", title='')
_ = ax2.legend(frameon=False, loc="upper right", title='')
_ = ax3.legend(frameon=False, loc="lower center", title='')
# 
# 保存图片
plt.savefig('plot_three_yaxis.png', dpi=300, bbox_inches='tight')

plt.show()
```
