## 3. Call Plot API

```python
fig, axs = plt.subplots(nrows=2, ncols=3, 
                        figsize=(7.2, 5),
                        gridspec_kw={
                            "width_ratios": [1,1.5,1]
                        })

plt.subplots_adjust(hspace=0.4, wspace=0.3)

for i in range(2):
    for j in range(3):
        txt = f'({chr(97+i*3+j)}) {i}, {j} plot'
        axs[i,j].set_title(txt, y=-0.3)

# plt.savefig('demo.png', dpi=300, bbox_inches='tight')

plt.show()
```
