# Insert Axes

## 1. Import matplotlib and numpy

```python
import matplotlib.pyplot as plt 
import numpy as np
from matplotlib import font_manager


plt.rcParams['font.size'] = 10

plt.rcParams['xtick.direction'] = 'in'
plt.rcParams['ytick.direction'] = 'in'
```

## 2. Prepare Data

```python
x = np.linspace(0, np.pi, 400)
f = 2*x**0.4 + np.sin(0.5*x) + 0.15*np.random.random(400)*np.sin(200*x)
```

## 3. Call Plot API

```python
fig, ax = plt.subplots(figsize=(3.5,3))  

ax.plot(x, f, lw=1.0, color='b', alpha=0.75, label='f data')
ax.set_xlabel("x/unit")
ax.set_ylabel("y/unit")

# subregion of the original image
x1, x2, y1, y2 = 1.8, 2.0, 3.2, 3.6  
axins = ax.inset_axes(
    [0.5, 0.1, 0.47, 0.47],
    xlim=(x1, x2), ylim=(y1, y2)) # , xticklabels=[], yticklabels=[]
axins.plot(x, f, lw=1.0, color='b', alpha=0.75, label='f data')
ax.indicate_inset_zoom(axins, edgecolor="black")


# plt.savefig('demo.png', dpi=300, bbox_inches='tight')

plt.show()
```
