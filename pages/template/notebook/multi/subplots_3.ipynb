## 3. Call Plot API

```python
from matplotlib import gridspec

fig = plt.figure(tight_layout=True, figsize=(6,8))
gs = gridspec.GridSpec(3, 2, width_ratios=[1, 2])

ax1 = fig.add_subplot(gs[0, :])
ax1.set_ylabel('YLabel0')
ax1.set_xlabel('XLabel0')
ax1.set_aspect(0.75)

for i in range(2):
    ax = fig.add_subplot(gs[1, i])
    ax.set_ylabel('YLabel1 %d' % i)
    ax.set_xlabel('XLabel1 %d' % i)

ax3 = fig.add_subplot(gs[2, :])
ax3.set_ylabel('YLabel0')
ax3.set_xlabel('XLabel0')
ax3.set_aspect(0.5)

# plt.savefig('demo.png', dpi=300, bbox_inches='tight')
plt.savefig('./cover/subplots_3.png', dpi=100, bbox_inches='tight')

plt.show()
```
