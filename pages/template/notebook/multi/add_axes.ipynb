# Add a image

## 1. Import matplotlib and numpy

```python

import matplotlib.pyplot as plt 
import numpy as np
from matplotlib import font_manager
import matplotlib.image as mpimg 


plt.rcParams['font.size'] = 10

plt.rcParams['xtick.direction'] = 'in'
plt.rcParams['ytick.direction'] = 'in'
```

## 2. Prepare Data

```python
img = mpimg.imread('./your_iamge_path.png')

x = [1, 2, 3, 4, 5, 6, 7]
y1 = [0.31, 0.37, 0.42, 0.41, 0.35, 0.32, 0.30]
y2 = [0.27, 0.31, 0.38, 0.44, 0.45, 0.39, 0.36]
```

## 3. Call Plot API

```python
fig, ax = plt.subplots(figsize=(3.5, 3))

_ = ax.plot(x, y1, color='r', marker='s', ls='-', label='data: y1')
_ = ax.plot(x, y2, color='b', marker='o', ls='-', label='data y2')
_ = ax.legend(frameon=False, loc=0, title='')

ax.set_xlim([0.7, 7.4])

ax.set_xlabel("x/unit")
ax.set_ylabel("y/unit")

ax.set_xticks(x)
ax.set_yticks([0.3, 0.35, 0.40])

# 在主图上添加一个子图，位置由 bbox_to_anchor 控制
subax = fig.add_axes([0.7, 0.65, 0.20, 0.20], zorder=1)
subax.axis("off")
subax.imshow(img, alpha=0.9)

# plt.savefig('demo.png', dpi=300, bbox_inches='tight')


plt.show()
```
